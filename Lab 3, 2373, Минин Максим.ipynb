### Метка классификации

В качестве целевой метки выступит атрибут POS, т.е. мы будем классифицировать игроков по игровым позициям, отталкиваясь от их статистики.

Позиции в датасете:

PG - разыгрывающий защитник
SG - атакующий защитник
SF - легкий форвард
PF - тяжелый форвард
C - центровой

### Оценка датасета

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy
import seaborn
```

```python
data = pd.read_csv("./dataset/nba.csv")
data.info()
```

Избавимся от записей с пустыми полями

```python
data = data.dropna()
```

Посмотрим сколько уникальных меток существует в датасете

```python
data['POS'].value_counts(normalize=True)
```

Видим, что классы, в принципе, сбалансированы.

## Классификация

```python
stat_columns = [col for col in data if col != 'POS' and col != 'Name']

X = data[stat_columns]
Y = data['POS']
```

Так как элементы атрибута POS - строки, конвертируем их. А также стандартизуем статистические данные

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

encoder = LabelEncoder()
Y = encoder.fit_transform(Y)

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

Поделим данные на тренировочные и тестовые

```python
from sklearn.model_selection import train_test_split

X_tr, X_t, Y_tr, Y_t = train_test_split(X, Y, stratify=Y)
```

## kNN

Сперва найдем оптимальный параметр k для kNN при помощи GridSearch

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

grid_params = {'n_neighbors': [i for i in range(1, 30+1)],
               'weights': ['uniform', 'distance'],
               'metric': ['minkowski', 'euclidean', 'manhattan']
              }

gs = GridSearchCV(knn, grid_params, scoring='accuracy', refit=True)

g_res = gs.fit(X_tr, Y_tr)

k = g_res.best_params_['n_neighbors']
metric = g_res.best_params_['metric']
weights = g_res.best_params_['weights']

g_res.best_params_
```

Лучшие параметры для kNN приведены выше. Теперь же, пользуясь этими значениями, проведем kNN-классификацию.

```python
knn = KNeighborsClassifier(n_neighbors=k, weights=weights).fit(X_tr, Y_tr)

predicted = knn.predict(X_t)
```

## Оценка качества классификации

Accuary оценка качества классификации:

```python
right = 0
wrong = 0

for i, pos in enumerate(Y_t):
    if predicted[i] == pos:
        right += 1
    else:
        wrong += 1
        
accuracy = right / (right + wrong)
print(accuracy)
```

Precision, Recall, F1 оценки качества:

```python
from yellowbrick.classifier import ClassificationReport

report = ClassificationReport(knn, classes=encoder.classes_)
report.fit(X_tr, Y_tr)
report.score(X_t, Y_t)
report.show();
```

ROC оценка качества:

```python
from yellowbrick.classifier import ROCAUC

report = ROCAUC(knn, classes=encoder.classes_)
report.fit(X_tr, Y_tr)
report.score(X_t, Y_t)
report.show();
```

## Дерево решений

Проведем классификацию

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc.fit(X_tr, Y_tr)

predicted = dtc.predict(X_t)
```

### Оценка качества классификаций

Accuracy оценка качества классификации

```python
right = 0
wrong = 0

for i, pos in enumerate(Y_t):
    if predicted[i] == pos:
        right += 1
    else:
        wrong += 1
        
accuracy = right / (right + wrong)
print(accuracy)
```

Precision, Recall, F1 оценки качества:

```python
report = ClassificationReport(dtc, classes=encoder.classes_)
report.fit(X_tr, Y_tr)
report.score(X_t, Y_t)
report.show();
```

ROC оценка классификации

```python
report = ROCAUC(dtc, classes=encoder.classes_)
report.fit(X_tr, Y_tr)
report.score(X_t, Y_t)
report.show();
```

## Анализ классификации

Можно сказать, что классификация, глобально, прошла умеренно успешно обоими методами. Видно, что модели тяжело дается распознавание игроков на позции SG (атакующий защитник) и PF (тяжелый форвард). Данный факт можно попытаться обосновать: 1) Говоря про SG, нужно сказать что, в действительности, игроков на данной позици действительно тяжело отличить чисто статистически, т.к. их самый больший показатель, зачастую - это PPG и AST. Притом эти же показатели зачастую больше у игроков на позициях SF и PG соответственно. 2) Говоря про PF, можно отметить, что данная позиция и позиция С - очень похожи. Игроки на этих позициях набирают примерно одинаковую статистику, т.к. их роли на паркете зачастую перекликаются. И при этом, PF способны набирать столько же PPG, как SF Резюмируя, можно предположить, что обе эти проблемы можно было бы уменьшить, если бы в датасете был еще один атрибут - рост игрока. Потому что тогда модель могла бы точнее различать PG и C, вследствие чего происходило бы меньше путаницы для игроков SG и PF.

Однако, модель хорошо распознает игроков на PG и C. Это, скорее всего, связано с тем, что игроки PG, зачастую, набирают больше всего APG и SPG, а игроки C выделяются в RPG и BPG.
